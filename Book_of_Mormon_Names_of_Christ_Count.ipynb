## Process Names Text

```python
import re
fa = re.findall
names = fa('\d+\. (.*?)$',namesRaw, re.M)
names = [i.lower() for i in names]
print(names[:5])
```

## Count Occurrences

```python
import time

counts = {}
for i in names:
  counts[i]=0
t1 = time.time()
for i in names:
  counts[i] = len(fa(i,bomText))
t2 = time.time()
print('Processed all names in {:.1f} milliseconds.'.format((t2-t1)*1000))

# Remove duplicates
for i in names:
  dups = [j for j in names if ((i in j) and (i != j))]
  if len(dups) > 0:
    dupsSum = sum([counts[j] for j in dups])
    counts[i] -= dupsSum
sortedCounts = sorted(counts.items(),key=lambda x: x[1],reverse=True)
for i in sortedCounts:
  print('{:>50} {:>6,}'.format(i[0],i[1]))

mentions = sum([i[1] for i in counts.items()])
print('Total mentions: {:,}\nVerses per mention: {:.1f}'.format(mentions,31102/mentions))
```
